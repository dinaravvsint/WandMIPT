### Récuperation des données 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = ".\\data\\"
file = "914-180324-152629-396.txt"

file_txt = PATH + file
data = np.genfromtxt(file_txt,delimiter='\t')

directs = ['T','AX','AY','AZ','HX','HY','HZ','MX','MY','MZ','Temp','Move']
```

### Format DataFrame de pandas 

```python
dat = pd.DataFrame(data[1:], columns=directs)
dat.head(10)
```

### Les graphes 

```python
timeline = dat['T']
num_signals = len(directs) - 3
fig, axes = plt.subplots(num_signals, figsize=(40,100), sharex=True)
plt.xticks(fontsize=40)

for i in range(num_signals):
    index = directs[i + 1]
    axes[i].plot(list(timeline),list(dat[index]), color='b', linewidth=2)
    axes[i].set_title(index+'(t)', fontdict={'fontsize':40})
plt.show()
```
